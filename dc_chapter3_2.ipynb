```python
import requests
from bs4 import BeautifulSoup
def get_book_title(isbn):
  url = 'http://www.yes24.com/Product/Search?domain=ALL&query={}'
  r = requests.get(url.format(isbn))
  soup = BeautifulSoup(r.text, 'html.parser')
  title = soup.find('a', attrs={'class':'gd_name'}).get_text()
  return title
```

```python
from pandas.core.internals.base import T
import re

def get_book_info(row):
  title = row['도서명']
  author = row['저자']
  pub = row['출판사']
  year = row['발행년도']

  url = 'http://www.yes24.com/Product/Search?domain=ALL&query={}'
  r = requests.get(url.format(row['ISBN']))
  soup = BeautifulSoup(r.text, 'html.parser')
  try:
    if pd.isna(title): #클래스 이름이 'gd_name'인 <a> 태그의 텍스트를 가져옴.
      title = soup.find('a', attrs={'class':'gd_name'}).get_text()
  except AttributeError:
    pass

  try:
    if pd.isna(author): #클래스 이름이 'info_auth'인 <span> 태그 아래 <a> 태그의 텍스트를 가져옴.
      authors = soup.find('span', attrs={'class':'info_auth'}).find_all('a')
      author_list = [auth.get_text() for auth in authors]
      author = ','.join(author_list)
  except AttributeError:
    pass

  try:
    if pd.isna(pub): #클래스 이름이 'info_pub'인 <span> 태그 아래 <a> 태그의 텍스트를 가져옴.
      pub = soup.find('span', attrs={'class':'info_pub'}).find('a').get_text()
  except AttributeError:
    pass

  try:
    if year == -1: #클래스 이름이 'info_pub'인 <span>의 택스트를 가져옴.
      year_str = soup.find('span', attrs={'class':'info_date'}).get_text()
      year = re.findall(r'\d{4}', year_str)[0]
  except AttributeError:
    pass

  return title, author, pub, year
```

```python
def data_fixing(ns_book4):
  """
  잘못된 값을 수정하거나 NaN을 채우는 함수
  :param ns_book4: data_cleaning() 함수에서 전처리된 데이터프레임
  """
  ns_book4 = ns_book4.astype({'도서권수':'int32', '대출권수':'int32'}) #도서권수와 대출권수를 int32로 바꿈.
  set_isbn_na_rows = ns_book4['세트 ISBN'].isna()
  ns_book4.loc[set_isbn_na_rows, '세트 ISBN'] = ''
  ns_book5 = ns_book4.replace({'발행년도':'.*(\d{4}).*'}, r'\1', regex=True) #발행년도 열에서 연도 네 자리를 추출하여 대체. 나머지 발행년도는 -1로 바꿈.
  unkown_year = ns_book5['발행년도'].str.contains('\D', na=True)
  ns_book5.loc[unkown_year, '발행년도'] = '-1'
  ns_book5 = ns_book5.astype({'발행년도': 'int32'}) #발행년도를 int32로 바꿈.
  dangun_yy_rows = ns_book5['발행년도'].gt(4000) 
  ns_book5.loc[dangun_yy_rows, '발행년도'] = ns_book5.loc[dangun_yy_rows, '발행년도'] - 2333 #4000년 이상인 경우 2333년을 뺌.
  dangun_year = ns_book5['발행년도'].gt(4000)
  ns_book5.loc[dangun_year, '발행년도'] = -1 #여전히 4000년 이상인 경우 -1로 바꿈.
  old_books = ns_book5['발행년도'].gt(0) & ns_book5['발행년도'].lt(1900) # 0~1999년 사이의 발행년도는 -1로 바꿈.
  ns_book5.loc[dangun_year, '발행년도'] = -1
  ns_rows = ns_book5['도서명'].isna() | ns_book5['저자'].isna() | ns_book5['출판사'].isna() | ns_book5['발행년도'].eq(-1)
  updated_sample = ns_book5[na_rows].apply(get_book_info, axis=1, result_type='expand') #yes24 도서 상세 페이지에서 누락된 정보를 채움.
  ns_book6 = ns_book5.dropna(subset=['도서명','저자','출판사'])
  ns_book6 = ns_book6[ns_book6['발행년도'] != -1] #도서명, 저자, 출판사가 NaN이거나 발해년도가 -1인 행을 삭제.

  return ns_book6
```
